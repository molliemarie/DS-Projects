###What is the probability that the tourist is at least 3 city blocks (as the crow flies) from Broadway and Broadway after 10 moves?

```python
NSteps = 10   # Number of steps in simulation.
blocks = 3    #Number of blocks we need to be outside of
#For the question 

count = 0
stat_bool = []
var = []
while count < 5000:
    
    xStart = 0   # x coordinate of starting location. Origin is at centre of square
    yStart = 0   # y coordinate of starting location. Origin is at centre of square
    x = xStart   # x coordinate of point.
    y = yStart   # y coordinate of point.
    xList = []   # List of the x coordinates of all points visited.
    yList = []   # List of the y coordinates of all points visited.

    
    for j in range(NSteps):
        new = random.randint(1,4)
        if new == 1:
            x += 1
        elif new == 2:
            y += 1
        elif new ==3 :
            x += -1
        else :
            y += -1
        xList.append(x)
        yList.append(y)
        
    if abs(xList[-1]) >=blocks:
        var = 1
    elif abs(yList[-1]) >=blocks:
        var = 1
    else:
        var = 0
    
    #In doing the above, I assumed that you wanted it to be at least 3 blocks away in either 
    #the x or y direction. However, if you had wanted to know if it were 3 walking blocks away,
    #it would have looked more like this:
    #if abs(xlist[-1]) + abs(ylist[-1]) >=3:
        #var = 1
    #else:
        #var = 0
        
    stat_bool.append(var)
    count +=1
```

```python
sum(stat_bool) / 5000
```
