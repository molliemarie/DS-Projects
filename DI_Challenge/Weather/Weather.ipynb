##San Francisco Weather, January 2015

Data from http://www.ncdc.noaa.gov/

####Meanings of columns

DAPR - Number of days included in the multiday precipitation total (MDPR)

WSF2 - Fastest 2-minute wind speed (tenths of meters per second)

WSF5 - Fastest 5-second wind speed (tenths of meters per second)

SNOW - Snowfall (mm)

WT03 - Thunder

WESF - Water equivalent of snowfall (tenths of mm)

PRCP - Precipitation (tenths of mm)

TOBS - Temperature at the time of observation (tenths of degrees C)

WT08 - Smoke or haze 

SNWD - Snow depth (mm)

WDF2 - Direction of fastest 2-minute wind (degrees)

AWND - Average daily wind speed (tenths of meters per second)

WDF5 - Direction of fastest 5-second wind (degrees)

PGTM - Peak gust time (hours and minutes, i.e., HHMM)

WT11 - High or damaging winds

WT01 - Fog, ice fog, or freezing fog (may include heavy fog)

TMAX - Maximum temperature (tenths of degrees C)

WESD - Water equivalent of snow on the ground (tenths of mm)

WT02 - Heavy fog or heaving freezing fog (not always distinguished from fog)

TMIN - Minimum temperature (tenths of degrees C)

MDPR - Multiday precipitation total (tenths of mm; use with DAPR and DWPR, if available)

WDMV - 24-hour wind movement (km)

###Closer Look at SF in Jan2015 

```python
tmax = SFAirport['TMAX'] / 10 #convert to degrees C
tmin = SFAirport['TMIN'] / 10 #convert to degrees C
rain = SFAirport['PRCP'] / 10 #convert to mm rainfall
wind = SFAirport['AWND']  #average daily wind speed in tenths of meters per second
fastest2wind = SFAirport['WSF2']  #fastest 2 minute wind speed in tenths of meters per second
fastest5wind = SFAirport['WSF5']  #fastest 5 minute wind speed in tenths of meters per second
fog = SFAirport['WT01'] # Fog, ice fog, or freezing fog (may include heavy fog)
smoke = SFAirport['WT08'] # smoke or haze
heavyFog = SFAirport['WT02'] # Heavy fog or heaving freezing fog (not always distinguished from fog)
```

```python
x = range(31)
plt.bar(x,list(tmax))
plt.ylabel('Max Temp (degrees C)')
plt.xlabel('Day in January 2015')
plt.title('Max Temp January 2015')
```

```python
plt.bar(x,list(tmin))
plt.ylabel('Min Temp (degrees C)')
plt.xlabel('Day in January 2015')
plt.title('Min Temp January 2015')
```

```python
plt.bar(x,list(rain))
plt.ylabel('Rain (mm)')
plt.xlabel('Day in January 2015')
plt.title('Rainfall in January 2015')
```

```python
plt.bar(x,list(wind))
plt.ylabel('Average Daily Wind Speed (tenths of meters per second)')
plt.xlabel('Day in January 2015')
plt.title('Average Wind Speed in January 2015')
```

```python
plt.bar(x,list(fastest2wind))
plt.ylabel('Fastest 2-Minute Wind Speed (tenths of meters per second)')
plt.xlabel('Day in January 2015')
plt.title('Fastest 2-Min Wind Speed in January 2015')
```

```python
plt.bar(x,list(fastest5wind))
plt.ylabel('Fastest 5-Minute Wind Speed (tenths of meters per second)')
plt.xlabel('Day in January 2015')
plt.title('Fastest 5-Min Wind Speed in January 2015')
plt.axis([0, 35, 0, 130])
```

```python
plt.bar(x,list(fog.replace(-9999, 0)))
plt.xlabel('Fog')
plt.title('Foggy Days in January 2015')
```

```python
plt.bar(x,list(heavyFog.replace(-9999, 0)))
plt.xlabel('Heavy Fog')
plt.title('Heavily Fogged Days in January 2015')
```

```python
plt.bar(x,list(smoke.replace(-9999, 0)))
plt.xlabel('Smoke')
plt.title('Smokey Days in January 2015')
```
